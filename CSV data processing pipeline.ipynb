```python
import pandas as pd
import os

INPUT_FOLDER = "data"
OUTPUT_FILE = "merged_output.csv"

FILL_NUMERIC = True
FILL_CATEGORICAL = True
DROP_ROWS_THRESHOLD = 0.5

def load_csv_files(folder_path):
    dataframes = []
    
    for file in os.listdir(folder_path):
        if file.endswith(".csv"):
            file_path = os.path.join(folder_path, file)
            print(f"Loading: {file}")
            df = pd.read_csv(file_path)
            dataframes.append(df)
    
    return dataframes

def clean_dataframe(df):
    threshold = int(len(df.columns) * DROP_ROWS_THRESHOLD)
    df = df.dropna(thresh=threshold)

    if FILL_NUMERIC:
        for col in df.select_dtypes(include=['number']).columns:
            df[col] = df[col].fillna(df[col].mean())

    if FILL_CATEGORICAL:
        for col in df.select_dtypes(include=['object']).columns:
            if not df[col].mode().empty:
                df[col] = df[col].fillna(df[col].mode()[0])

    return df

def merge_dataframes(dfs):
    merged_df = pd.concat(dfs, ignore_index=True)
    return merged_df

def main():
    print("Starting data pipeline...")

    dfs = load_csv_files(INPUT_FOLDER)

    if not dfs:
        print("No CSV files found!")
        return

    cleaned_dfs = [clean_dataframe(df) for df in dfs]

    merged_df = merge_dataframes(cleaned_dfs)

    merged_df.to_csv(OUTPUT_FILE, index=False)

    print(f"Data merged and saved to {OUTPUT_FILE}")
    print(f"Final shape: {merged_df.shape}")

if __name__ == "__main__":
    main()
```
